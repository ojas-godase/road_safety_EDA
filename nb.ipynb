# EDA

```python
import pandas as pd

# 1️⃣ Total number of accidents in India across all causes
total_accidents = df_india.filter(like="Total Accidents").sum().sum()

# 2️⃣ Total number of deaths caused by these accidents
total_deaths = df_india.filter(like="Death").sum().sum()

# 3️⃣ Comparison between different causes
causes = ["Over-Speeding", "Drunken Driving", "Driving on Wrong side", "Jumping Red Light", "Use of Mobile Phone", "Others"]
accidents_by_cause = {cause: df_india.filter(like=cause).filter(like="Total Accidents").sum().sum() for cause in causes}
deaths_by_cause = {cause: df_india.filter(like=cause).filter(like="Death").sum().sum() for cause in causes}

# 4️⃣ Which type of highway (NHAI, State PWD, or Other departments) has the most accidents?
accidents_by_highway = {
    "NHAI": df_india.filter(like="NHAI").filter(like="Total Accidents").sum().sum(),
    "State PWD": df_india.filter(like="State PWD").filter(like="Total Accidents").sum().sum(),
    "Other Departments": df_india.filter(like="Other department").filter(like="Total Accidents").sum().sum()
}

# 5️⃣ Which type of accident is the deadliest? (highest death rate)
death_rates = {cause: (deaths_by_cause[cause] / accidents_by_cause[cause]) if accidents_by_cause[cause] > 0 else 0 for cause in causes}
deadliest_cause = max(death_rates, key=death_rates.get)

# 6️⃣ Find the most common cause of accidents
most_common_cause = max(accidents_by_cause, key=accidents_by_cause.get)

# 7️⃣ Rank accident causes by number of deaths
deaths_by_cause_sorted = dict(sorted(deaths_by_cause.items(), key=lambda x: x[1], reverse=True))

# 8️⃣ Rank highways by total accidents
accidents_by_highway_sorted = dict(sorted(accidents_by_highway.items(), key=lambda x: x[1], reverse=True))

# Print results
print("🚦 Road Accident Analysis in India 🚦")
print("="*50)
print(f"✅ Total Accidents in India: {total_accidents}")
print(f"✅ Total Deaths in India: {total_deaths}")
print("="*50)

print("\n📊 Accidents by Cause:")
for cause, count in accidents_by_cause.items():
    print(f"{cause}: {count}")

print("\n💀 Deaths by Cause (Ranked):")
for rank, (cause, deaths) in enumerate(deaths_by_cause_sorted.items(), start=1):
    print(f"{rank}. {cause}: {deaths} deaths")

print("\n🛣️ Accidents by Highway Type (Ranked):")
for rank, (highway, accidents) in enumerate(accidents_by_highway_sorted.items(), start=1):
    print(f"{rank}. {highway}: {accidents} accidents")

print(f"\n☠️ The deadliest cause of accidents is: {deadliest_cause} with a death rate of {death_rates[deadliest_cause]:.2f}")
print(f"\n🚗 The most common cause of accidents is: {most_common_cause} with {accidents_by_cause[most_common_cause]} accidents")
print("="*50)
```

```python
# 1️⃣ Over-Speeding Accidents by Highway Type
over_speeding_highway = {
    "NHAI": df_india.filter(like="Over-Speeding").filter(like="NHAI").filter(like="Total Accidents").sum().sum(),
    "State PWD": df_india.filter(like="Over-Speeding").filter(like="State PWD").filter(like="Total Accidents").sum().sum(),
    "Other Departments": df_india.filter(like="Over-Speeding").filter(like="Other department").filter(like="Total Accidents").sum().sum(),
}

# 2️⃣ Fatality Rates: Drunken Driving vs Over-Speeding
drunk_driving_death_rate = deaths_by_cause["Drunken Driving"] / accidents_by_cause["Drunken Driving"] if accidents_by_cause["Drunken Driving"] > 0 else 0
over_speeding_death_rate = deaths_by_cause["Over-Speeding"] / accidents_by_cause["Over-Speeding"] if accidents_by_cause["Over-Speeding"] > 0 else 0

# 3️⃣ Percentage of Accidents Due to Jumping Red Lights & Mobile Phone Use
red_light_accident_pct = (accidents_by_cause["Jumping Red Light"] / total_accidents) * 100
mobile_phone_accident_pct = (accidents_by_cause["Use of Mobile Phone"] / total_accidents) * 100

# 4️⃣ Ranking Red Light Jumping & Mobile Phone Use by Severity (Death Rate)
red_light_death_rate = deaths_by_cause["Jumping Red Light"] / accidents_by_cause["Jumping Red Light"] if accidents_by_cause["Jumping Red Light"] > 0 else 0
mobile_phone_death_rate = deaths_by_cause["Use of Mobile Phone"] / accidents_by_cause["Use of Mobile Phone"] if accidents_by_cause["Use of Mobile Phone"] > 0 else 0

# 5️⃣ Fatality Rate by Highway Type
deaths_by_highway = {
    "NHAI": df_india.filter(like="NHAI").filter(like="Death").sum().sum(),
    "State PWD": df_india.filter(like="State PWD").filter(like="Death").sum().sum(),
    "Other Departments": df_india.filter(like="Other department").filter(like="Death").sum().sum(),
}
fatality_rate_by_highway = {k: (deaths_by_highway[k] / accidents_by_highway[k]) if accidents_by_highway[k] > 0 else 0 for k in deaths_by_highway}

# 6️⃣ NHAI vs. State PWD Accident & Death Rates
nhai_accident_rate = (accidents_by_highway["NHAI"] / total_accidents) * 100
state_pwd_accident_rate = (accidents_by_highway["State PWD"] / total_accidents) * 100
nhai_fatality_rate = fatality_rate_by_highway["NHAI"]
state_pwd_fatality_rate = fatality_rate_by_highway["State PWD"]

# 📢 Print Results
print("\n🚗 Over-Speeding Accidents by Highway Type:")
for highway, count in over_speeding_highway.items():
    print(f"{highway}: {count} accidents")

print(f"\n⚖️ Fatality Rates Comparison:")
print(f"Drunken Driving Death Rate: {drunk_driving_death_rate:.2f}")
print(f"Over-Speeding Death Rate: {over_speeding_death_rate:.2f}")

print(f"\n🚦 Percentage of Accidents Due to:")
print(f"Jumping Red Light: {red_light_accident_pct:.2f}%")
print(f"Mobile Phone Use: {mobile_phone_accident_pct:.2f}%")

print(f"\n☠️ Severity Ranking by Death Rate:")
print(f"Jumping Red Light Death Rate: {red_light_death_rate:.2f}")
print(f"Mobile Phone Use Death Rate: {mobile_phone_death_rate:.2f}")

print(f"\n🛣️ Fatality Rate by Highway Type:")
for highway, rate in fatality_rate_by_highway.items():
    print(f"{highway}: {rate:.2f}")

print(f"\n📊 NHAI vs. State PWD:")
print(f"NHAI Accident Share: {nhai_accident_rate:.2f}%")
print(f"State PWD Accident Share: {state_pwd_accident_rate:.2f}%")
print(f"NHAI Fatality Rate: {nhai_fatality_rate:.2f}")
print(f"State PWD Fatality Rate: {state_pwd_fatality_rate:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1️⃣ Accident Severity Rate (Deaths per Accident for each cause)
causes = ["Over-Speeding", "Drunken Driving", "Driving on Wrong side", "Jumping Red Light", "Use of Mobile Phone", "Others"]
accidents_by_cause = {cause: df_india.filter(like=cause).filter(like="Total Accidents").sum().sum() for cause in causes}
deaths_by_cause = {cause: df_india.filter(like=cause).filter(like="Death").sum().sum() for cause in causes}

severity_rate = {cause: (deaths_by_cause[cause] / accidents_by_cause[cause]) if accidents_by_cause[cause] > 0 else 0 for cause in causes}

# 2️⃣ Cause-wise Contribution (%)
total_accidents = sum(accidents_by_cause.values())
cause_contribution = {cause: (accidents_by_cause[cause] / total_accidents) * 100 for cause in causes}

# 🔹 Insights Printing
print("🚦 **Accident Severity Rate (Deaths per Accident) by Cause:**")
for cause, rate in severity_rate.items():
    print(f"{cause}: {rate:.2f} deaths per accident")

deadliest_cause = max(severity_rate, key=severity_rate.get)
print(f"\n☠️ **The deadliest cause of accidents is:** {deadliest_cause} with a severity rate of {severity_rate[deadliest_cause]:.2f} deaths per accident.")

print("\n📊 **Cause-wise Contribution to Total Accidents (%):**")
for cause, percentage in cause_contribution.items():
    print(f"{cause}: {percentage:.2f}% of total accidents")

most_common_cause = max(cause_contribution, key=cause_contribution.get)
print(f"\n🚗 **The most common cause of accidents is:** {most_common_cause}, contributing {cause_contribution[most_common_cause]:.2f}% to total accidents.")
```

# Linear Regression 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

# Assume df2 is already loaded
df2.fillna(df2.mean(), inplace=True)  # Handle missing values

# Aggregate accident types across all road departments
X = pd.DataFrame()
X['Total Drunken Driving Accidents'] = df2.filter(like='Drunken ').filter(like='Total Accidents').sum(axis=1)
X['Total Over-Speeding Accidents'] = df2.filter(like='Over-Speeding').filter(like='Total Accidents').sum(axis=1)
X['Total Mobile Phone Use Accidents'] = df2.filter(like='Use of Mobile Phone').filter(like='Total Accidents').sum(axis=1)

# Aggregate all deaths
y = df2.filter(like='Death').sum(axis=1)
y.name = "Total Deaths"

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train Linear Regression Model
model = LinearRegression()
model.fit(X_scaled, y)

# Get Regression Coefficients
coefficients = pd.DataFrame(model.coef_, X.columns, columns=['Coefficient'])
coefficients = coefficients.sort_values(by='Coefficient', ascending=False)

# Print coefficients
print("Regression Coefficients:\n", coefficients)

# Visualizing Feature Importance
plt.figure(figsize=(10, 5))
sns.barplot(x=coefficients.index, y=coefficients['Coefficient'])
plt.xticks(rotation=45)
plt.title("Impact of Drunken Driving, Over-Speeding, and Mobile Phone Use on Deaths")
plt.xlabel("Accident Causes")
plt.ylabel("Coefficient Value")
plt.show()

# Heatmap of correlations
plt.figure(figsize=(8, 5))
sns.heatmap(X.join(y).corr(), cmap="coolwarm", annot=True, fmt=".2f")
plt.title("Correlation Heatmap of Selected Accidents & Deaths")
plt.show()
```

# Correlation Analysis of Accidents & Deaths

This heatmap shows the correlation between different types of accidents and total deaths.  

- **Over-Speeding Accidents & Deaths (0.99)** → Very strong correlation, meaning more over-speeding accidents lead to more deaths.  
- **Drunken Driving Accidents & Deaths (0.96)** → Also strongly correlated, confirming its fatal risk.  
- **Mobile Phone Use Accidents & Deaths (0.89)** → A bit lower than the others but still significant.  
- **Over-Speeding & Other Accidents (0.94 - 0.95)** → Over-speeding is highly correlated with both drunken driving and mobile phone use accidents, possibly because reckless behavior overlaps.  

🔹 **Over-speeding is the biggest factor in road deaths, followed closely by drunken driving and mobile phone use.**
